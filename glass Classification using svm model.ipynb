```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=df.iloc[:,:-1]
y=df['Type']
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.15,random_state=5)
```

```python
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_val=scale.transform(x_val)
x_test=scale.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 1, 10]
}
svm_model = svm.SVC()
grid_search = GridSearchCV(svm_model, param_grid, cv=5)
grid_search.fit(x_train, y_train)
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
svm_model=svm.SVC(C= 1, gamma= 0.1, kernel= 'rbf')
svm_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_train=svm_model.predict(x_train)
y_pred_val=svm_model.predict(x_val)
print('accuracy of x_train = ',accuracy_score(y_train,y_pred_train))
print('accuracy of x_val = ',accuracy_score(y_val,y_pred_val))
```

```python
y_pred_test=svm_model.predict(x_test)
print('accuracy of x_test = ',accuracy_score(y_test,y_pred_test))
```

```python
sns.scatterplot(x=y_test,y=y_pred_test,cmap='Reds')
plt.xlabel('y_test')
plt.ylabel('y_predict')
plt.show()
```

```python
index = np.arange(len(y_test))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(index, y_test, c='blue', label='Actual')
ax.scatter(index, y_pred_test, c='red', label='Predicted')
ax.set_xlabel('Index')
ax.set_ylabel('Values')
ax.set_zlabel('Target')
ax.legend()
plt.show()
```
